#PCA

```python
import numpy as np
import matplotlib . pyplot as plt
from PIL import Image

# Load image and convert to grayscale
image = Image.open ("/content/drive/MyDrive/Sem 6/AI/Week 1/naruto.png").convert ("L")

# Convert to NumPy array
image_array = np.array(image)
print(image_array)

# Get image dimensions
height , width = image_array.shape

# Reshape the image into a 2D array where each row represents a row of pixels
# This allows PCA to process each row as an observation and find patterns across pixel intensities
data = image_array.reshape(height , width )

plt.imshow(data, cmap='gray')
plt.show()
```

##Standardize the Data:

```python
# Compute the mean of each column ( feature )
mean = np . mean ( data , axis =0)
# Subtract mean to center the data
centered_data = data - mean
```

##Compute Covariance Matrix

```python
# Compute covariance matrix
cov_matrix = np . cov ( centered_data , rowvar = False )
```

##Compute Eigenvalues & Eigenvectors:

```python
# Compute eigenvalues and eigenvectors
eigenvalues , eigenvectors = np . linalg . eigh ( cov_matrix )
# Sort eigenvalues and corresponding eigenvectors in descending order
sorted_indices = np . argsort ( eigenvalues ) [:: -1]
eigenvalues = eigenvalues [ sorted_indices ]
eigenvectors = eigenvectors [: , sorted_indices ]
```

##Picking the Principal Components:

```python
#Plot the explained variance ratio
explained_variance_ratio = eigenvalues / np .sum ( eigenvalues )
plt . plot ( np . cumsum ( explained_variance_ratio ) )
plt . title (" Cumulative Explained Variance ")
plt . xlabel (" Number of Components ")
plt . ylabel (" Cumulative Explained Variance ")
plt . grid ( True )
plt . show ()
```

##Select Top k Principal Component :

```python
k = 50 # Choose k principal components
components = eigenvectors [: , : k ]
```

##Compress the Image

```python
# Project the data onto the principal components
compressed_data = np . dot ( centered_data , components )
```

##Reconstruct (Decompress) the Image

```python
# Reconstruct the image from compressed data
decompressed_data = np . dot ( compressed_data , components . T ) + mean
```

##Final Visualization:

```python
plt.figure(figsize=(12,6))
# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image_array, cmap="grey")
plt.title("OriginalImage")
plt.axis("off")
# Compressed Representation
plt.subplot(1, 3, 2)
plt.imshow(compressed_data,cmap="grey", aspect="auto")
plt.title(f"CompressedImage(k={k})")
plt.axis("off")
#Decompressed Image
plt.subplot(1, 3, 3)
plt.imshow(decompressed_data, cmap="grey")
plt.title("DecompressedImage")
plt.axis("off")
plt.tight_layout()
plt.show()
```
